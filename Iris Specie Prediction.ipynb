# Iris Specie Prediction

```python
# (Setosa, Versicolour, and Virginica)
```

```python
# import library
# นำเข้า pandas สำหรับอ่านไฟล์ csv และจัดการข้อมูลรูปแบบตาราง
import pandas as pd
```

```python
# load dataset
# อ่านข้อมูลแล้วเก็บในตัวแปร df
df = pd.read_csv("Iris.csv")
```

```python
# explore dataset (eda)
# แสดง 5 แถวแรกของข้อมูล
df.head()
```

```python
# type of species
# ดูสายพันธ์ต่างๆ
df['Species'].unique()
```

```python
# data preprocessing
# เช็คข้อมูลที่หายไป แต่ละคอลัมน์จะบอก NaN
df.isna().sum()
```

```python
# features and labels
# x - features ตัวเลขที่ใช้ทำนาย
# y - labels คำตอบที่เป็นชนิดของดอกไม้
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']
```

```python
# import train_test_split
# ฟังก์ชันแบ่งข้อมูล
from sklearn.model_selection import train_test_split
```

```python
# final preprocessing
# X_train, y_train = 70% ใช้สอนโมเดล
# X_test, y_test = 30% ใช้ทดสอบความแม่นยำ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# import LogisticRegression
# ใช้สำหรับ classification
from sklearn.linear_model import LogisticRegression
```

```python
# create model
# สร้างโมเดลเปล่า ๆ 1 ตัว
model = LogisticRegression()
```

```python
# train model
# ex.1 ถ้ากลีบดอกสั้น -> setosa
# ex.2 ถ้ายาวมาก -> virginica
model.fit(X_train, y_train)
```

```python
# predict model
# ทำนายข้อมูล test เก็บไว้ใน predictions
predictions = model.predict(X_test)
```

```python
# import evaluate tools
# นำเข้าเครื่องมือประเมินผลโมเดล
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# check model accuracy
# แสดงรายงานสรุป precision, recall, f1-score, support
print(classification_report(y_test, predictions))
```

```python
# guess amount
# แสดง matrix ว่า ทายถูกกี่ตัว ทายผิดเป็นอะไร
print(confusion_matrix(y_test, predictions))
```

```python
# example test model
# สร้างข้อมูลดอกไม้ใหม่ 1 ตัวอย่าง เพื่อทดลองทำนาย แล้วแปลงเป็น df
test_plant = {
    'SepalLengthCm': 5.1,
    'SepalWidthCm': 3.5,
    'PetalLengthCm': 1.4,
    'PetalWidthCm': 0.2
}

plant_df = pd.DataFrame([test_plant])
plant_df
```

```python
# predict example test model
# ทำนายข้อมูลที่เราสร้าง
model.predict(plant_df)
```
